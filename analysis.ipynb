```python
import pandas as pd
import numpy as np


data=pd.read_csv('htseq.100.50.clean.csv')

data['prc_change']=data['cov50']/data['cov100']
data['cov100_category'] = np.where(data['cov100']<50, '<50', '>50')

data.head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

g=sns.set_style("white")
g=sns.set_context("talk")
sns.violinplot(x="cov100_category", y="prc_change", data=data)
#g.set(xlabel='Number of reads', ylabel='Percentage decrease')
g=sns.despine()
plt.savefig("Violinplot.pdf", bbox_inches='tight')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

g=sns.set_style("white")
g=sns.set_context("talk")
g=sns.regplot(x="cov100", y="prc_change", data=data,color="black")
g.set(xlabel='cov100', ylabel='prc_change')
g=sns.despine()
#plt.show(g)
plt.savefig("100_vs_50.pdf", bbox_inches='tight')
```
