```python
import pandas as pd


data=pd.read_csv('htseq.100.50.clean.csv')




data.head()
```

```python
data2=data.loc[data['cov100'] <50]
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from scipy import stats

g=sns.set_style("white")
g=sns.set_context("talk")
g=sns.regplot(x="cov100", y="cov50", data=data2,color="black")
g.set(xlabel='cov100', ylabel='cov50')
g=sns.despine()
#plt.show(g)
plt.savefig("100_vs_50.pdf", bbox_inches='tight')
```
